# 1) Contexte

```python
def f(x) :
    return( 2 + np.cos(4*x)+ x + x**2 - np.exp(x)+ 0.3*np.sin(12*x))
```

```python
x=np.linspace(0,1,1000)
plt.plot(x,f(x))
```

```python
def estimation_L(x,f):
    somme=0
    for i in range(x.shape[0]-1):
        der=(f(x[i+1])-f(x[i]))/(x[i+1]-x[i])
        somme+=np.sqrt(1+ (der**2))
    return(somme/x.shape[0])
```

```python
L=estimation_L(x,f)
```

```python
print(L)
```

# 2) Métamodèle par processus gaussien

```python
def cov_matern3(sigma, l, x):
    return(sigma*(1+sqrt(6)*np.abs(x)/l)*np.exp(-sqrt(6)*np.abs(x)/l))
```

```python
def matrice_cov(fct,n,sigma,l):
    pts=np.linspace(0,1,n)
    cov=np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            val=pts[i]-pts[j]
            cov[i,j]=fct(sigma,l,val)
    return(cov)


def moy_condi(x,sigma,l,n,fct):
    Y_obs=f(x)
    Y_est=[]
    mat=matrice_cov(fct,x.shape[0],sigma,l)
    x2=np.linspace(0, 1, n)
    for i in range(n):
        mat2=np.zeros(x.shape[0])

        for j in range(x.shape[0]):
            mat2[j]=fct(sigma,l,x2[i]-x[j])
        Y_est+=[np.dot(np.dot(mat2.T, np.linalg.inv(mat)), Y_obs)]
    return(Y_est)
```

```python
x_2=np.array([0,0.2,0.4,0.6,0.8,1])
sig,l=0.2,0.4
```

```python
Y_hat=moy_condi(x_2,sig,l,1000,cov_matern3)
```

```python
plt.plot(x_2,f(x_2),'-o')
plt.plot(np.linspace(0,1,1000),Y_hat)
```

```python
def var_condi(x,sigma,l,n,fct):
    Y1=[]
    x2=np.linspace(0, 1, n)
    mat=matrice_cov(fct,x.shape[0],sigma,l)
    for i in range(n):
        mat2=np.zeros(x.shape[0])
        for j in range(x.shape[0]):
            mat2[j]=fct(sigma,l,x2[i]-x[j])
        Y1.append(sigma-np.dot(np.dot(mat2.T, np.linalg.inv(mat)), mat2))
    return Y1
```

```python
var_est=var_condi(x_2,sig,l,1000,cov_matern3)
int_sup = Y_hat+1.96*np.sqrt(var_est)
int_inf = Y_hat-1.96*np.sqrt(var_est)
```

```python
plt.plot(np.linspace(0, 1, 1000), Y_hat, 'b', label='Moyenne conditionnelle')
plt.plot(np.linspace(0, 1, 1000), int_inf, 'r', label='Intervalle de confiance à $95\%$')
plt.plot(np.linspace(0, 1, 1000), int_sup, 'r')
plt.plot(x_2, f(x_2), 'bo', label="Points d'observations")
plt.legend()
plt.grid()
plt.title('Approximation de f(x)')
plt.xlabel('x')
plt.ylabel('y')
```

```python
def Lmin(f_esti):
    taille=len(f_esti)
    x=np.linspace(0,1,taille)
    lmin=0
    h=1/taille
    for i in range (taille-1):
        if i+2==taille:
            f2=np.sqrt(1+((f_esti[i+1]-f_esti[i+1])/h)**2)
        else :
            f2=np.sqrt(1+((f_esti[i+2]-f_esti[i+1])/h)**2)
        f1=np.sqrt(1+((f_esti[i+1]-f_esti[i])/h)**2)
        lmin+=h*(f1+f2)/2
    return lmin
```

```python
print("Approximation de la quantité d'intérêt :", Lmin(Y_hat))
```

# 3 - Evaluation de la longueur minimale par simulations conditionnelles

```python
def process(sig,m):
    n=len(sig)
    x=np.random.rand(n)
    p=[]
    for i in range(n):
        p+=[(sig[i])*x[i]+m[i]]
    return(p)
```

```python
a=np.linspace(0, 1, 1000)
plt.plot(a, Y_hat, 'b', label='Moyenne conditionnelle')
plt.plot(a, int_inf, 'r', label='Intervalle de confiance à $95\%$')
plt.plot(a, int_sup, 'r')
plt.plot(x_2, f(x_2), 'bo', label="Points d'observations")
for i in range(5):
    plt.plot(a,process(var_est,Y_hat))
plt.legend()
plt.grid()
plt.title('Approximation de f(x)')
plt.xlabel('x')
plt.ylabel('y')
```

```python
simu_gauss=np.zeros((200,1000))
Lmin_1=[]
for i in range(200):
    simu_gauss[i]=process(var_est,Y_hat)
    Lmin_1+=[Lmin(simu_gauss[i])]
```

```python
print('Estimation de L_min : %f'%(np.mean(Lmin_1)))
print("Intervalle de confiance[",Lmin_1[int(0.05*len(Lmin_1))],",", Lmin_1[int(0.95*len(Lmin_1))],"]")
```
